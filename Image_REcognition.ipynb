```python
import cv2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image  = cv2.imread('/content/WhatsApp Image 2024-03-18 at 21.07.06.jpeg')
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")


plt.subplot(1,2,2)

plt.imshow(gray_image,cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.show()
```

```python
#Medical imaging:Enhance contrast
enhanced_image = cv2.equalizeHist(gray_image)

#Remote Sensing : Edge detection
edges = cv2.Canny(gray_image,100,200)

#Biometrics:- Face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray_image,scaleFactor=1.3,minNeighbors=5)
#industrail inspection: image segmentation
ret,threshold_image = cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY)

#Display results
plt.figure(figsize=(15,10))
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB))
plt.title('Enhanced Image')
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(edges,cmap='gray')
plt.title('Edges')
plt.axis('off')

plt.subplot(2,3,3)
for(x,y,w,h) in faces:
  cv2.rectangle(image, (x,y), (x+w, y+h), (255, 0, 0), 2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Face Detection')
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(threshold_image,cmap='gray')
plt.title('Threshold Image')
plt.axis('off')

plt.show()
```

```python
# Noise

demonised_image = cv2.fastNlMeansDenoisingColored(image,None,10,10,7,21)



resized_image = cv2.resize(image,(500,500))

blurred_image = cv2.GaussianBlur(image,(5,5),0)

num_labels,labels = cv2.connectedComponents(threshold_image)

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(demonised_image,cv2.COLOR_BGR2RGB))
plt.title("Demonize Image")
plt.axis("Off")

plt.subplot(2,3,2)
plt.imshow(cv2.cvtColor(resized_image,cv2.COLOR_BGR2RGB))
plt.title("Resized Image")
plt.axis("Off")


plt.subplot(2,3,3)
plt.imshow(cv2.cvtColor(blurred_image,cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.axis("Off")

plt.subplot(2,3,4)
plt.imshow(labels.astype(np.uint8)*255,cmap='gray')
plt.title("Connected Components")
plt.axis("Off")


plt.show()
```
